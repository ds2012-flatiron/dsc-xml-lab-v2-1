## XML

```python
import xml.etree.ElementTree as ET
```

### Create an XML tree and retrieve the root tag.

```python
#Your code here
tree = ET.parse('nyc_2001_campaign_finance.xml')
root = tree.getroot()
print(f'XML root tag: "{root.tag}", XML root attrib: "{root.attrib}"')
```

```python
tag = root[0].tag
attr = root[0].attrib
print(tag, '\n', attr)
rootChild = root[0]
for child in rootChild:
    print(child.tag, child.attrib, child.text)
    for child in child:
        print(child.tag, child.attrib, child.text)
```

### How many direct descendents does the root tag have?

```python
#Your code here
numRtDirDescendents = 0
for child in root:
    print(child.tag, child.attrib)
    numRtDirDescendents += 1
print(f'Number of Root Direct Descendents: {numRtDirDescendents}')
```

### How many different types of tags are there within the entire XML file?

```python
# Your code here
dctTags = dict()
rootChild = root[0]
for child in rootChild:
    print(child.tag, child.text, child.attrib)
    print('--------child of root child---------')
    if child.tag not in dctTags.keys():
        dctTags[child.tag] = 1
    else:
        dctTags[child.tag] += 1
    for child in child:
        print(child.tag, child.text,child.attrib)        
        print('--------child of child of root child---------')
        if child.tag not in dctTags.keys():
            dctTags[child.tag] = 1
        else:
            dctTags[child.tag] += 1

```

### Create a DataFrame listing the number of each type of tag. 
Sort the DataFrame in descending order by the tag count. The first entry should demonstrate there are 286 row tags in the XML file.

```python
#Your code here
df = pd.DataFrame({
    'tag': dctTags.keys(),
    'counts' : dctTags.values()
})
df.sort_values('counts', inplace=True, ascending=False)
df
```
